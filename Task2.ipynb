```python
import pandas as pd

data = pd.read_csv("C:\\Users\\harsh\\Downloads\\01.Data Cleaning and Preprocessing.csv")

data2 = data.fillna(value=0)
```

```python
import numpy as np
from scipy import stats 


data2.drop(['Observation'], axis=1, inplace=True)
data2.columns
```

```python
Q1 = data2.quantile(0.25)
Q3 = data2.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
data2=data2[~((data2<(Q1-1.5*IQR))|(data2>(Q3+1.5*IQR))).any(axis=1)]
data2
```
